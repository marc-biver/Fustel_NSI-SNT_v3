## Exercice 1 : édition et exécution d'une cellule de code

Un notebook jupyter consiste en **cellules** individuelles pouvant contenir du **code** Python, du **texte**, ou des **graphiques**. Commençons par les cellules de type Code.

1. Placez le curseur dans la 1ère cellule <code>Entrée[ ]:</code> ci-dessous puis frappez `<enter>`. Que se passe-t-il ? Rien d'autre que l'insertion d'un saut de ligne... C'est normal, car dans une cellule de type "Code" (c'est-à-dire préfixée par <code>Entrée[ ]:</code>) on est par défaut en mode **édition de code** Python !


Pour **exécuter le code** Python d'une cellule, il faut en effet frapper `<maj-entrée>` (ou cliquer sur le bouton <button class='fa fa-step-forward icon-step-forward'></button>, ou encore utiliser le menu `Cellule > Exécuter les cellules et sélectionner la suivante`). Essayez !

```python
15 + 24
```

```python
'Hello tout le monde !'
```

Une ligne <code>Sortie[ ]:</code> devrait être apparue en dessous de la 1ère expression, affichant le résultat de celle-ci ! Le curseur est alors automatiquement passé dans la cellule d'en-dessous que vous pouvez exécuter à son tour.

_Notez qu'il est aussi possible de frapper `<ctrl-entrée>` (identique à `Cellule > Exécuter les cellules `). Dans ce cas, le curseur **reste dans la cellule** après l'exécution du code._

2. Exécutez maintenant le code de la cellule ci-dessous.

```python
a = 6     # affectation de la valeur 6 à la variable a
b = 10    # affectation de la valeur 10 à la variable b

print('La somme de ces 2 nombres vaut :', a + b)
```

## Quelques raccourcis clavier bien utiles à noter dans un coin...

On retrouve tous les raccourcis dans le menu « Aide », mais il est bon de retenir ce qui suit.  <a name="cop"></a> 

Il faut d'abord cliquer sur la gauche de la cellule (partie blanche) puis, pour  
**1. Copier :** taper <code>c</code>   
Couper : <code>x</code>  
Coller : <code>v</code>  
Supprimer :  <code>d,d</code> (c'est-à-dire taper deux fois sur d)  


**2. Insérer une cellule avant :** <code>a</code> (comme above) <a name="ins"></a>  
Insérer une cellule après : <code>b</code>(comme below)  


**3. Mettre en Markdown, éditer,**   <a name="mark"></a>  
Pour mettre la cellule en Markdown : <code>m</code>  
Pour mettre la cellule en code : <code>y</code>  
Pour éditer une cellule en Markdown : <code>enter</code> (entrée)  


**4. Pour exécuter une cellule :** <code>shift + enter</code>   <a name="exe"></a> 


**5. Annuler la dernière action :**  
<code>z</code> pour annuler une suppression de cellule  
<code>Ctrl + z</code>pour annuler ce que l'on vient de taper  (seulement en mode édition)
