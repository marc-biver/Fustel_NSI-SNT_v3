# Exercice 1 : évolution dans le temps du prix d'un article

1. On considère le code Python suivant :

```python
def afficheTousLesPrix(a,n):
    for i in range(1,n+1):
        print(a)
        a = a + 3.5
    return a

afficheTousLesPrix(50,5)
```

2. Cette fonction est-elle satisfaisantes pour étudier l'évolution du prix de l'article au cours du temps ?

**Cette fonction ne permet de stocker en mémoire QUE la valeur du prix au bout de N années. Si je veux obtenir la valeur du prix après M (différent de N) années, je dois ré-exécuter la fonction.**

On considère à présent le code Python suivant :

```python
def tousLesPrix(a,n):
    prix = [a]
    for i in range(1,n+1):
        a = a + 3.5
        prix.append(a)
    return prix

tousLesPrix(50,5)
```

# Exercice 3 : nombre de licenciés d'un club de triathlon

Un club de triathlon compte $1500$ licenciés. Chaque année, il perd $5\%$ de ses licenciés et enregistre $400$ nouveaux inscrits.

Compléter la fonction suivante pour qu'elle renvoie une liste contenant l'évolution du nombre de licenciés sur un nombre `nbAnnees` d'années.

```python
def nbLicenciesTriathlon(nbAnnees):
    listeNbLicencies = [1500]
    for i in range(nbAnnees):
        listeNbLicencies.append(int(listeNbLicencies[i] * 0.95 + 400)) # on calcule un nombre entier de licenciés !
    return listeNbLicencies

# Vous noterez ici que chaque itération de la boucle calcule la prochaine valeur
# sur la base de la valeur qu'elle vient de calculer (par le biais de listeNbLicencies[i])
```

Les quatre cellules suivantes permettent de tester votre fonction. Elles doivent renvoyer `True` lors de leur exécution.

```python
nbLicenciesTriathlon(0) == [1500]
```

```python
nbLicenciesTriathlon(1) == [1500, 1825]
```

```python
nbLicenciesTriathlon(4) == [1500, 1825, 2133, 2426, 2704]
```

```python
nbLicenciesTriathlon(8) == [1500, 1825, 2133, 2426, 2704, 2968, 3219, 3458, 3685]
```

# Exercice 4 : échange d'éléments dans une liste

La fonction `echange` ci-dessous prend en paramètre une liste non vide `L` et deux entiers `i` et `j`, et renvoie la liste `L` dans laquelle les éléments d'indices `i` et `j` ont été échangés.

Avant d'effectuer l'échange des deux éléments, la fonction effectue trois tests d'assertion :
 - le premier vérifie que `L` est bien une liste non vide ;
 - le deuxième vérifie que l'entier `i` est bien un nombre compatible avec un indice de `L` ;
 - le troisième vérifie que l'entier `j` est bien un nombre compatible avec un indice de `L`.

Compléter cette fonction `echange`.

```python
def echange(L,i,j):
    temp = L[i]
    L[i] = L[j]
    L[j] = temp
    return L
```

Les trois cellules suivantes permettent de tester votre fonction. Elles doivent renvoyer `True` lors de leur exécution.

```python
echange([8,9],0,1) == [9,8]
```

```python
echange([1,2,3,5,4,6,7],3,4) == [1,2,3,4,5,6,7]
```

```python
echange(['A','C','B','D'],1,2) == ['A','B','C','D']
```

# Exercice 6 : insertion d'élément dans une liste

1. La fonction `insertionDebutListe` ci-dessous prend en paramètre une liste `L` et une donnée `element`, et renvoie une liste dans laquelle `element` a été inséré au début de `L` à l'aide de l'opérateur de concaténation.

Compléter cette fonction.

```python
def insertionDebutListe(L,element):
    return [element] + L
```

Les trois cellules ci-dessous sont un jeu de tests pour tester votre fonction; elles doivent toutes renvoyer True.

```python
insertionDebutListe([],5) == [5]
```

```python
insertionDebutListe([4],0) == [0,4]
```

```python
insertionDebutListe([2,1,5],-3) == [-3,2,1,5]
```

2. En s'aidant de la fonction `insertionDebutListe` écrite ci-dessus, écrire une fonction `insertionFinListe` qui permet d'insérer un élément à la fin d'une liste.

```python
# Ecrire ici le code de la fonction insertionFinListe
def insertionFinListe(L,element):
    return L + [element]
    # Note: un simple "append" aurait aussi fonctionné.... "L.append(element)"
```

Les trois cellules ci-dessous sont un jeu de tests pour tester votre fonction; elles doivent toutes renvoyer True.

```python
insertionFinListe([],5) == [5]
```

```python
insertionFinListe([4],0) == [4,0]
```

```python
insertionFinListe([2,1,5],-3) == [2,1,5,-3]
```

3. Reprenez votre fonction `insertionDebutListe` ci-dessus et complétez-la pour qu'elle n'insère l'élément demandé _que_ s'il n'est pas déjà présent dans la liste (ainsi, un appel à `insertionDebutListe([1, 2],0)` renverra `[0, 1, 2]` tandis qu'un appel à `insertionDebutListe([1, 2],1)` renverra `[1, 2]` - donc la liste inchangée).

```python
# Ecrire ici le code de la fonction insertionDebutListe
def insertionDebutListe(L,element):
    if element in L: # On utilise l'opérateur "in" qui permet de faire un test d'appartenance.
        return L
    else:
        return [element] + L
```

Les trois cellules ci-dessous sont un jeu de tests pour tester votre fonction; elles doivent toutes renvoyer True.

```python
insertionDebutListe([1, 2],9) == [9, 1, 2]
```

```python
insertionDebutListe([1, 2],1) == [1, 2]
```

```python
insertionDebutListe([1, 2],2) == [1, 2]
```
