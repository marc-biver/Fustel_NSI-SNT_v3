Ce que vous venez d'utiliser s'appelle en réalité des **"balises"** - et le langage HTML en compte des dizaines de différentes. En règle générale, pour une balise appelée NOM_BALISE, on l'ouvre avec la syntaxe `<NOM_BALISE>` et on la ferme avec la syntaxe `</NOM_BALISE>` en mettant entre l'ouverture et la fermeture le texte auquel on veut appliquer l'effet de la balise - comme on a fait ci-dessus avec `h1`.
<br><br><br>
Voici quelques exemples de balises - testez-les dans votre page et répondez aux questions ci-dessous:
